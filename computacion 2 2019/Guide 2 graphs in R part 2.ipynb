## Introducción a gráficos
### Histograma
Probemos haciendo un histograma usando un vector.

```r
# Define con 5 valores
suvs <- c(4,4,6,6,16)

# Crea un histograma:
hist(suvs)
```

Usemos los datos anteriormente creados en la guía anterior para hacer un histograma:

```r
cars <- c(1, 3, 6, 4, 9)
trucks <- c(2, 5, 4, 5, 12)
autos_data=data.frame(cars,trucks)
```

Concatenamos los datos en 3 vectores y creamos el histograma con el eje y entre los valores 0-10:

```r
autos <- c(autos_data$cars, autos_data$trucks, 
   autos_data$suvs)

hist(autos, col="lightblue", ylim=c(0,10))
```

Ahora cambiamos los "breaks":

```r
autos <- c(autos_data$cars, autos_data$trucks, 
   autos_data$suvs)

max_num <- max(autos)

hist(autos, col=heat.colors(max_num), breaks=max_num, 
   xlim=c(0,max_num), right=F, main="Autos Histogram", las=1)
```

Ahora probemos cambiando las opciones gráficas:

```r
autos <- c(autos_data$cars, autos_data$trucks, 
   autos_data$suvs)
max_num <- max(autos)

brk <- c(0,3,4,5,6,10,16)

hist(autos, col=heat.colors(length(brk)), breaks=brk, 
   xlim=c(0,max_num), right=F, main="Probability Density", 
   las=1, cex.axis=0.8, freq=F)
```

Ahora graficaremos 1000 valores aleatorios con una distribución log-normal.

```r
r <- rlnorm(1000)

hist(r)
```

Ya que la distribución log-normal se ve mejor con ejes log, podemos usar la función plot con puntos para mostrar la distribución

```r
h <- hist(r, plot=F, breaks=c(seq(0,max(r)+1, .1)))
plot(h$counts, log="xy", pch=20, col="blue",
	main="Distribución Log-normal",xlab="Value", ylab="Frequency")
```

### Gráficos de torta
Empecemos con algo simple:

```r
cars <- c(1, 3, 6, 4, 9)
pie(cars)
```

Ahora agreguemos unas etiquetas y un titulo:

```r
pie(cars, main="Cars", col=rainbow(length(cars)),
   labels=c("Lun","Mar","Mie","Jue","Vie"))
```

Ahora cambiemos los colores, etiquetas usando porcentajes, y creemos una leyenda:

```r
colors <- c("white","grey70","grey90","grey50","black")

car_labels <- round(cars/sum(cars) * 100, 1)
car_labels <- paste(car_labels, "%", sep="")

pie(cars, main="Cars", col=colors, labels=car_labels,
   cex=0.8)

legend(1.5, 0.5, c("Mon","Tue","Wed","Thu","Fri"), cex=0.8, 
   fill=colors)
```

### Gráficos de punto
Empecemos con los datos de los autos:

```r
dotchart(t(autos_data))
```

Podemos crear uno de estos gráficos un poco más coloridos

```r
dotchart(t(autos_data), color=c("red","blue","darkgreen"),
   main="Dotchart de Autos", cex=0.8)
```

### Bonus
Este ejemplo muestra todos los 25 símbolos que puedes usar para hacer puntos en tus gráficos:

```r
# Crea un gráfico vacío
plot(1, 1, xlim=c(1,5.5), ylim=c(0,7), type="n", ann=FALSE)

# Plot de dígitos 0-4 con color en tamaño creciente
text(1:5, rep(6,5), labels=c(0:4), cex=1:5, col=1:5)

# Plot de símbolos 0-4 con color en tamaño creciente
points(1:5, rep(5,5), cex=1:5, col=1:5, pch=0:4)
text((1:5)+0.4, rep(5,5), cex=0.6, (0:4))

# Plot de símbolos 5-9 con etiquetas
points(1:5, rep(4,5), cex=2, pch=(5:9))
text((1:5)+0.4, rep(4,5), cex=0.6, (5:9))

# Plot de símbolos 10-14 con etiquetas
points(1:5, rep(3,5), cex=2, pch=(10:14))
text((1:5)+0.4, rep(3,5), cex=0.6, (10:14))

# Plot de símbolos 15-19 con etiquetas
points(1:5, rep(2,5), cex=2, pch=(15:19))
text((1:5)+0.4, rep(2,5), cex=0.6, (15:19))

# Plot de símbolos 20-25 con etiquetas
points((1:6)*0.8+0.2, rep(1,6), cex=2, pch=(20:25))
text((1:6)*0.8+0.5, rep(1,6), cex=0.6, (20:25))
```
