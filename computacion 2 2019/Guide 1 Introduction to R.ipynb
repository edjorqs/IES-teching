Teniendo datos, es sumamente fácil poder hacer un histograma. Por ejemplo, a continuación crearemos un vector con distribución normal estándar de largo 1.000:

```r
x=rnorm(1000)
```

Y ahora haremos el histograma:

```r
hist(x)
```

```r
hist(x, prob=TRUE)
lines(density(x), # gráfico de densidad (sin asumir una distribución de probabilidad en particular)
 lwd = 2, # grosor de la línea
 col = "red")
```

R tiene similitudes con Python, pero tiene ventajas al tratarse de un software estadístico. Puedes crear un dataframe con la función `data.frame()`, en donde puedes insertar las columnas que quieras:

```r
x=1:100
y=2*x+1
df=data.frame(x,y)
head(df)
```

En R también existen las listas, como ya habíamos mencionado en clase, para ingresar a un elemento de un vector, basta con usar los corchetes a la derecha junto con el nombre del vector. Para acceder a un elemento de una lista, puedes  hacerlo mediante corchetes dobles:

```r
A=list(head(df),x,y,z)
cat("Esto es A:")
A
cat("Esto es el primer elemento de A:")
A[[1]]
```

Crear una función es sumamente sencillo, análogo a lo que vimos en Python, también se usa un `return()`:

```r
saludo <- function(hola){
    if(hola=="hola"){
        return(cat("hola"))
    }
    else{
        return(cat("chao"))
    }
}
saludo("hola")
```
