## 1. Regresión lineal simple

Ya vimos de qué trata una regresión lineal simple, ahora veamos cómo modelizar esto en R.

Veamos unos datos de ejemplo:

````r
#Altura
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)
#Peso
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)

# Usando lm()
relation <- lm(y~x)

print(relation)
````

Ahora obtengamos un resumen de esta relación:

````r
print(summary(relation))
````

## Predicción

La sintaxis básica para predecir según un modelo, es la siguiente:

```
predict(objeto, datos_nuevos)
```

Lo siguiente es a descripción de los parámetros usados:
- **objeto** es la fórmula que ya está creada usando la función `lm()`.
- **datos_nuevos** es el vector que contiene los datos nuevos para ser usados por la variable predictora.

## Predecir pesos de nuevas personas:

````r
# El vector de predicción
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)

# El vector de respuesta
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)

# Aplicando la función lm()
relation <- lm(y~x)

# Encuentra pesos de una persona con altura de 170.
a <- data.frame(x = 170)
result <-  predict(relation,a)
print(result)
````

## Visualicemos esto gráficamente:

````r
# Crea el predictor como una variable respuesta
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)
relation <- lm(y~x)

# Démosle un nombre al archivo del gráfico
png(file = "linearregression.png")

# Grafica la regresión
plot(y,x,col = "blue",main = "Height & Weight Regression",
abline(lm(x~y)),cex = 1.3,pch = 16,xlab = "Weight in Kg",ylab = "Height in cm")

# Save the file.
dev.off()
````

Si graficamos por separado, tendremos algo como:

````r
plot(y,x,col = "blue",main = "Height & Weight Regression",
abline(lm(x~y)),cex = 1.3,pch = 16,xlab = "Weight in Kg",ylab = "Height in cm")
````
